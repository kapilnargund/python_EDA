```python
# 1. Read Data and assign proper header values

import pandas as pd
df_2021 = pd.read_excel("Fortune-Global-500-List-2021.xlsx")
df_2022 = pd.read_excel("Fortune-500-Global-List-2022.xlsx")

df_2021 = df_2021.iloc[6:]
df_2021.drop(columns=df_2021.columns[0], axis=1, inplace=True)

df_2021 = df_2021.rename(columns=df_2021.iloc[0]).drop(df_2021.index[0]).reset_index(drop=True)
df_2021
```

```python
# 2. Replace new line to space in column names

for i in df_2021.columns:
    if "\n" in i:
        df_2021.rename(columns = {i:i.replace("\n","")}, inplace = True)
    
df_2021
```

```python
# 3. Show summary statistics

df_2021.describe()
```

```python
# 4. Display Country with most number of companies are present in list

print(df_2021['Country'].value_counts())
```

```python
# 5. Create Column Revenue 2021 and calculate its value : revenue_2021=revenue_2022+(revenue_2022*revenue change)

df_2022 = df_2022.iloc[6:]
df_2022.drop(columns=df_2022.columns[0], axis=1, inplace=True)

df_2022 = df_2022.rename(columns=df_2022.iloc[0]).drop(df_2022.index[0]).reset_index(drop=True)
df_2022

for i in df_2022.columns:
    if "\n" in i:
        df_2022.rename(columns = {i:i.replace("\n","")}, inplace = True)
    
df_2022['revenue_2021'] = df_2022['Revenues($millions)'] + (df_2022['Revenues($millions)']*df_2022['Revenue Change'])
df_2022
```

```python
# 6. Multiply Change(Revenue,Profit) values by 100 

df_2021['Revenue Change pct'] = df_2021['Revenue Change']*100
df_2021['Profit Change pct'] = df_2021['Profit Change']*100

df_2021
```

```python
# 7. Create column Revenue per Employee : Revenue/ Employee

df_2021['rev_per_emp'] = df_2021['Revenues($millions)'] / df_2021['Number of Employees']
df_2021
```

```python
# 8.  Filter Data where revenue >100000 and Revenue/ per Employee >5

for i in range(len(df_2021)):
    if df_2021['Revenues($millions)'][i] > 100000 and df_2021['rev_per_emp'][i] > 5:
        print(df_2021.iloc[i])
```

```python
# 9. Create Continent column from Country 

import pycountry_convert as pc

def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

# # Example
# country_name = 'Great Britain'
# print(country_to_continent(country_name))

cont_li = []
for i in range(len(df_2021)):
    try:
        if df_2021['Country'][i] == 'Britain':
            cont_li.append(country_to_continent('Great '+df_2021['Country'][i]))
        elif df_2021['Country'][i] == 'france':
            cont_li.append(country_to_continent('France'))
        else:
            cont_li.append(country_to_continent(df_2021['Country'][i]))
    except Exception as e:
        print(e)
        print(df_2021['Country'][i])
        
df_2021['Continent'] = cont_li

df_2021
```

```python
# 10. Display top 3 companies from each continent by revenue

gk = df_2021.groupby('Continent')
  
# Let's print the first entries
# in all the groups formed.
gk.first().sort_values('Revenues($millions)', ascending=False).head(3)
```
